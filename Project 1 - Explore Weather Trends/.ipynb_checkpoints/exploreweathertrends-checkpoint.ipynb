## Load data from CSV files

Here we import python libraries used to calculate the moving average and also libraries to plot graphs. Also, the df DataFrame loads data from city_data.csv downloaded previously; gdf DataFrame loads the global data.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

cdpath = 'C://Users//rafae//Desktop//Udacity//Project 1 - Explore Weather Trends//city_data.csv'
df = pd.read_csv(cdpath, header = 0)

gdpath = 'C://Users//rafae//Desktop//Udacity//Project 1 - Explore Weather Trends//global_data.csv'
gdf = pd.read_csv(gdpath, header = 0)
```

The cities - randomly chosen - are added to the array below. 

```python
cities = ['Dakar-Senegal', 'London-United Kingdom', 'Sydney-Australia','Oslo-Norway']
```

## Moving Average

As the dataset presents the temperature variation by year, and we are only interested to see the weather trends, a moving average per <b>decade</b> is added to the dataset so we can smooth the visualisation. 

```python
df['moving_average'] = df.groupby(['city','country'], as_index=False)['avg_temp'].rolling(10).mean().reset_index(0,drop=True).dropna()
gdf['moving_average'] = gdf['avg_temp'].rolling(10).mean().reset_index(0,drop=True).dropna()
```

Where:
    - groupby: groups the rows by city and country so we can apply the moving average correctly.
    - rolling: specify the size of the window, it calculates the average of the past ten records.
    - mean: after determining the moving average window size, we estimate the mean itself
    - dropna: ignore records where the moving average is missing

```python
fig, ax = plt.subplots()

# Add city lines
for city in cities:
    ax.plot(df[df['city']+'-'+df['country'] == city]['year'],df[df['city']+'-'+df['country'] == city]['moving_average'],label=city)

# Add global average line
ax.plot(gdf['year'],gdf['moving_average'],'c:',label='Global Average')

# Position the legend box out of the graph area
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

#resize the graph
fig.set_size_inches(18.5, 10.5, forward=True)

#set the axes labels
plt.xlabel('Year')
plt.ylabel('Temperature in Degrees')

# Highlight the Industrial revolution period
plt.axvspan(1760, 1840, color='red', alpha=0.1)

# Begining of cars' mass production
plt.axvspan(1900, 1940, color='yellow', alpha=0.1)


plt.show()
```
